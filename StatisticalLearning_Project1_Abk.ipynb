```python
import numpy as np
import pandas as pd
```

## Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
LaLigaSet = pd.read_csv('Laliga.csv')
LaLigaSet.replace('-', 0, inplace=True)
headers = LaLigaSet.iloc[0]
LaLiga = pd.DataFrame(LaLigaSet.values[1:], columns=headers)
LaLiga.head()
```

## Print all the teams which have started playing between 1930-1980.

```python
# LaLiga.Debut[3][0:4]
LaLigaBtwn = pd.DataFrame(LaLiga.loc[(LaLiga.Debut >= '1930') & (LaLiga.Debut <= '1980')])
LaLigaBtwn[['Team','Debut']]
```

## Print the list of teams which came Top 5 in terms of points

```python
# LaLigaBtwn.sort_values(by='Points', axis=0)
LaLigaP = LaLiga.astype({'Points': int, 'GoalsFor': int, 'GoalsAgainst': int, 'GamesWon': int, 'GamesPlayed': int, 'BestPosition': int})
LaLigaP.sort_values('Points', ascending=False, inplace=True)
LaLigaP[['Team', 'Points']].head()
```

## Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
def Goal_diff_count():
    LaLigaP['Goal_Differences'] = (LaLigaP['GoalsFor'] - LaLigaP['GoalsAgainst'])
    print('{} \n'.format(LaLigaP[['Team', 'Goal_Differences']]))
    
    if (LaLigaP.Goal_Differences.max() > np.absolute(LaLigaP.Goal_Differences.min())):
        print('Maximum Goal Difference is: {}'.format(LaLigaP['Goal_Differences'].max()))
        print('{} \n\n'.format(LaLigaP[LaLigaP.Goal_Differences == LaLigaP['Goal_Differences'].max()].Team))
    else:
        print('Maximum Goal Difference is: {} \n'.format(np.absolute(LaLigaP.Goal_Differences.min())))
        print('{}\n\n'.format(LaLigaP[LaLigaP.Goal_Differences == LaLigaP['Goal_Differences'].min()].Team))

    print('Minimum Goal Difference is: {}'.format(np.absolute(LaLigaP.Goal_Differences).min()))
    print('{}\n\n'.format(LaLigaP.loc[(LaLigaP.Goal_Differences == np.absolute(LaLigaP.Goal_Differences).min()) | (LaLigaP.Goal_Differences == -np.absolute(LaLigaP.Goal_Differences).min())].Team))


Goal_diff_count()

# Note: Assuming Max goal difference could be either positive or negative. 
# and min goal difference is closest to zero or zero
```

## Create a new column with name “Winning Percent” and append it to the data set

```python
LaLigaP.head()
LaLigaP["Winning Percent"] =  (LaLigaP.GamesWon / LaLigaP.GamesPlayed)*100
LaLigaP.sort_values("Winning Percent", ascending=False, inplace=True)
LaLigaP[["Team","Winning Percent"]].head(5)
```

## Group teams based on their “Best position” and print the sum of their points for all positions

```python
LaLigaB = LaLiga.astype({"BestPosition": int, "Points": int})
LaLigaB.sort_values("BestPosition")
LaLigaB.groupby(by=["BestPosition"], as_index=False).agg({'Points':'sum'})
```
